# Now I am Defining the agents

```python
# Define agents with specific roles and goals in cybersecurity
vulnerability_analyst = Agent(
    role='Vulnerability Analyst',
    goal='Identify and assess security vulnerabilities in software systems, after looking at some statistics from your monitoring tools like wazuh etc',
    backstory="""You are a skilled analyst at a cybersecurity firm, focused on finding weaknesses in software that attackers could exploit.
    You work for the companies Security Operations Team which is known as the SOC.
    You have about 10 things that you research on, which you research on a quaterly basis.Malware Ransomware Breaches Insider threats/privilege misuse Supply chain attacks Phishing Denial of service attacks Cyber-espionage.
    Your report is focused on those elements  """,
    verbose=True,
    allow_delegation=False,
    tools=[search_tool]
)


Graph_plotting_agent = Agent(
    role='Graph plotting agent',
    goal='Create graphs and a graph that will be used to show the trends in Phishing, Malware Ransomware Breaches Insider threats/privilege misuse Supply chain attacks Phishing Denial of service attacks Cyber-espionage ',
    backstory="""For the past 30 years You are responsible for creating detailed line graphs and bar hraphs that The Security report writer will use.
    you are responsible for helping to create visual elements to the report so that The stake holders can see some visual Metrics
    also you are responsible for creating simple pngs that can be added to the report as visual images  """,
    verbose=True,
    allow_delegation=True
)

security_report_writer = Agent(
    role='Security Report Writer',
  goal='Document and report on identified security vulnerabilities, and write an understandable report that you will save to a pdf file, then after that I want you to take the report and summarise it into a simple report at the end of the detailed report and make it as simple as a 9th grader to read, You will describe each thing you found and you will add an image for each explanation',
    backstory="""for the last 30 years, You are responsible for creating detailed reports on security findings, translating technical details into clear, actionable information for clients.
    You are also able to and write an understandable report that you will save to a pdf file, then after that I want you to take the report and summarise it into a simple report at the end of the detailed report and make it as simple as a 9th grader to read, You will describe each thing you found and you will add an image for each explanation""",
    verbose=True,
    allow_delegation=True
)

```

# Now I  am Defining the tasks that they will do

```python
# Create tasks for each agent
task_detect = Task(
    description="Scan the system and identify potential security vulnerabilities.",
    expected_output="List of vulnerabilities detected with details.Malware Ransomware Breaches Insider threats/privilege misuse Supply chain attacks Phishing Denial of service attacks Cyber-espionage",
    agent=vulnerability_analyst
)



task_Visual = Task(
    description="Write a comprehensive report on the vulnerabilities found, including potential impacts and mitigation strategies.",
    expected_output="Detailed security report ready for stakeholders.",
    agent=Graph_plotting_agent
)



task_report = Task(
    description="Write a comprehensive report on the vulnerabilities found, including potential impacts and mitigation strategies.",
    expected_output="Detailed security report with visual graphs and images that you will get from Graph_plotting_agent ready for stakeholders.with a simple 9th grader summary at the end and then I will be able to have a beautiful report ",
    agent=security_report_writer
)
```

**# Now I am Initiating the sequential process, it can either be heirachy, ow sequential**

```python
# Instantiate the crew with a sequential process
crew = Crew(
    agents=[vulnerability_analyst, Graph_plotting_agent, security_report_writer],
    tasks=[task_detect, task_Visual, task_report],
    verbose=2  # Higher verbosity for detailed logging
)
```

Now I am going to execute and print the tasks

```python
# Assuming 'crew' is properly defined and initialized
result = crew.kickoff()

# Display the result in the console
print("######################")
print(result)

# Save the result to a text file
with open("result.txt", "a") as file:  # "a" stands for "append mode"
    file.write("######################\n")
    file.write(str(result) + "\n")

```
